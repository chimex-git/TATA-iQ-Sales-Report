```python
import pandas as pd
import numpy as np
```

```python
# load the data

df = pd.read_csv('Online Retail Data Set.csv', encoding='ISO-8859-1', low_memory=False)
df
```

```python
df.dtypes
```

```python
df.info
```

```python
# replace either "?" or ".." that might be in the dataset so that the dropna function could recognise it as a missing value

df.replace("?", np.NaN, inplace=True)
```

```python
# drop missing rows from the dataframe

dropped_missing_rows = df.dropna(subset =["Description","CustomerID"], inplace=True)
```

```python
# check for missing values

df.isnull().sum()
```

```python
# drop all duplicates

df.drop_duplicates(inplace=True)
```

```python
# check for duplicates

df.duplicated().sum()
```

```python
df["Quantity"].tolist()
```

```python
# To check for quantities that are less than 1

low_quantity = df[df["Quantity"]<1]["Quantity"].tolist()
print(low_quantity)
```

```python
# To check for unit price that are less than 1

low_UnitPrice = df[df["UnitPrice"]<1]["UnitPrice"].tolist()
print(low_UnitPrice)
```

```python
# Keep only rows where Quantity is at least 1 AND Price is 0 or more
df_cleaned = df[(df['Quantity'] >= 1) & (df['UnitPrice'] >= 0)]
display(df_cleaned)
```

```python
# Export to a new CSV file
df.to_csv('Cleaned_Online_Retail_Data.csv', index=False, encoding='utf-8')
```
